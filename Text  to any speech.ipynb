```python
import os
from IPython.display import Audio, display, FileLink

VOICE_NAME='D'
text='Hello from this tutorial, I hope you enjoy it'

# Generate with your own voice
voice_samples, conditioning_latents = load_voice(VOICE_NAME)
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents)

# Save the generated voice to a wav file
output_filename = f'generated-{VOICE_NAME}.wav'
torchaudio.save(output_filename, gen.squeeze(0).cpu(), 24000)

# Display an option for downloading the generated voice
display(Audio(output_filename))
display(FileLink(output_filename))
```
